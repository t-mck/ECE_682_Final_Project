```python
# Import necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

import pandas as pd
import glob

# Get a list of all CSV files in the folder
csv_files = glob.glob("select_users/*.csv")

# Initialize an empty list to hold the DataFrames
df_list = []

# Loop through each CSV file and load it into a DataFrame, then append it to the list
for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)
```

```python
df_list[0]
```

```python
# Loop through each DataFrame in the list and drop the specified columns
for i in range(len(df_list)):
    df_list[i] = df_list[i].drop(['Unnamed: 0', 'name', 'address', 'city', 'state', 'categories'], axis=1)
```

```python
df_list[0]
```

```python
import pandas as pd
import numpy as np

# Loop through each DataFrame in the list and shuffle the rows
for i in range(len(df_list)):
    df_list[i] = df_list[i].sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_list[0]
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
for i in range(len(df_list)):
    # Get the first DataFrame in the list
    df = df_list[i]

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.1, random_state=50)

    # Initialize a random forest regression model
    model = RandomForestRegressor(n_estimators=500, random_state=50)

    # Train the model on the training data
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    
    print("mae ", mae, " mse ",mse, " rmse ", rmse, " for user number ", i)

    # Test the model on the testing data and get the R-squared score
    #r2_score = model.score(X_test, y_test)

    #print("R-squared score:", r2_score, "for user number ", i)
```
