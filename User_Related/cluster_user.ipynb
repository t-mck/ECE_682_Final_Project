```python
for col, vals  in uniques.items(): 
    if vals == {0,1}: 
        business_feats[col] = (business_feats[col]) * 2 - 1
    if vals.issuperset({False, np.nan}): 
        business_feats[col] = (business_feats[col]) * 2 - 1
        business_feats[col].fillna(0, inplace = True)
```

```python
uniques = {}
for column in business_feats.columns: 
    uniques[column] =  set(business_feats[column].unique())
for col, vals  in uniques.items(): 
    if vals != {0, 1, -1} or vals != {1, -1}: 
        print(col) 
        print(vals)
```

```python
new_user_data = pd.concat([new_user_data, pd.DataFrame(0, index = new_user_data.index, columns = business_feats.columns)], axis = 1)
```

```python
def get_cell_scores(row):
    
    num_bus = 0
    for ranking in [1,2,3,4,5]:
        if row[str(ranking)]== row[str(ranking)]:
            buses =  eval(row[str(ranking)])
            for business in buses: 
                num_bus += 1
                ## Option 1: (ranking-3) * [-1,1] + (ranking-3) * [-1,0,1] 
                row[5:] +=  (ranking - 3) * business_feats.loc[business]
    row[5:] /= num_bus
    return row
```

```python
def add_num_bus(row): 
    num_bus = 0
    for ranking in [1,2,3,4,5]:
        if row[str(ranking)]== row[str(ranking)]:
            buses =  eval(row[str(ranking)])
            num_bus += len(buses)
    row['num_businesses'] = num_bus
    return row
```

```python
user_data_feats = new_user_data.progress_apply(get_cell_scores, axis = 1)
```

```python
user_data_feats2 = user_data_feats.progress_apply(add_num_bus, axis = 1)
```
